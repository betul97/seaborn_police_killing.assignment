**INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
# with seaborn
#default
```

```python
city_d=kill.city.value_counts().head(12)
city_d
```

```python
city_d_i=city_d.index
city_d_i
```

```python
city_d_v=city_d.values
city_d_v
```

## 2: Plot 12 most dangerous cities

```python
#seaborn barplot showing
plt.figure(figsize=(8,6))
sns.barplot(x=city_d_i,y=city_d_v,data=kill)
plt.title('Most Dangerous Cities',color='blue',fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

```python
# with seaborn barplot

```

```python
#countplot
plt.figure(figsize=(8,6))
sns.countplot(data=kill,x='city',order=city_d_i);
plt.title("Most dangerous cities",color="blue",fontsize=15)
plt.xticks(rotation=45);
plt.show()
```

```python
# with seaborn countplot
```

## 3: Plot 20 most dangerous states

```python
state=kill.state.value_counts().head(20)
state
```

```python
state_i=state.index
state_i
```

```python
state_v=state.values
state_v
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x=state_i,y=state_v,data=kill)
plt.title("Most Dangerous States",color="blue",fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

```python
# with seaborn barplot
```

```python
fig, ax =plt.subplots(figsize=(7,5))
sns.countplot(x="state", data=kill,
             order=state.index,
             )
ax.set_xticklabels(state.index, rotation = 45)
ax.set_title('Most Dangerous States',color = 'blue',fontsize=15)
plt.tight_layout()
```
